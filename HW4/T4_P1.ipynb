```python

import numpy as np
import matplotlib.pyplot as plt

x = np.array([-3, -2, -1, 0, 1, 2, 3])
y = np.array([1, 1, -1, 1, -1, 1, 1])

def phi(x):
    return (x, -8/3*x**2 + 2/3*x**4)

print(phi(x))

phi_x = phi(x[y==1])
plt.plot(phi_x[0], phi_x[1], 'ob', )
phi_x = phi(x[y==-1])
plt.plot(phi_x[0], phi_x[1], 'xr')
plt.plot(x, [-1 for _ in range(len(x))], '-k')
plt.xlabel('phi(x)_1')
plt.ylabel('phi(x)_2')
plt.savefig("svm1.png")
plt.show()
```

```python

def phi2(x):
    return (x, -4*x**2 + 1/2*x**4)
phi_x = phi2(x[y==1])
plt.plot(phi_x[0], phi_x[1], 'ob')
phi_x = phi2(x[y==-1])
plt.plot(phi_x[0], phi_x[1], 'xr')
plt.plot(x, [-6 for _ in range(len(x))], '-k')
plt.xlabel('phi(x)_1')
plt.ylabel('phi(x)_2')
plt.savefig("svm2.png")
plt.show()
```
